### Funciones lambda

* Funciones anónimas
* Definir funciones por el usuario con una notación más simplificada
* Conviene cuando son tareas rápidas-cortas
* No conviente cuando quieres implementar tareas robustas
* No puedes colocar documentación

Esqueleto para crear una función en Python

```python
def nombreFuncion(param_1, param_2, ....):
  # tarea a realizar
  # instruccion de retorno (opcional)
  return
```

Esqueleto para funciones lambda

```python
nombreFuncion = lambda param_1, param_2, ...: tarea a realizar de la funcion - <return>
```

Función anónima:

* Función lambda sin nombre

```python
lambda param_1, param_2, ...: tarea a realizar de la funcion - <return>
```

````python
def f(x):
  """
  Documentacion de la funcion:
  * ¿que hace tu funcion?
  * ¿que necesita? Parametros
  * ¿Que regresa?
  * Reglas
  """
  return x ** 2
````

````python
# lambda x: x ** 2 ---> funcion anonima
fl = lambda x: x ** 2
````

````python
print(f(4))
print(fl(4))
````

### Función apply de pandas 🐼

* Aplica una función, idealmente anónima, a cada elemento/fila de una columna

````python
# 1. Acceder a la columna de interes
# Fecha_Inicio
df_bank["Fecha_Inicio"] = df_bank["Fecha_Inicio"].apply(lambda x: parser.parse(x))

# fillna(X): te rellena los valores faltantes de la columna de interes
# con el valor X
df_bank["Fecha_Fin"] = df_bank["Fecha_Fin"].fillna("1900-01-01")
df_bank["Fecha_Fin"] = df_bank["Fecha_Fin"].apply(lambda x: parser.parse(x))
````

````python
df_bank["Plazo"] = df_bank["Plazo"].fillna(0)
````

````python
# te da el tipo de dato de tus columnas
# cuantos nulos por columnas tienes
df_bank.info()
````

````python
# conversion de tipos de datos astype(nuevoTipoDeDatos)
df_bank["Plazo"] = df_bank["Plazo"].astype(int)
````

````python
df_bank.head(8)
````

````python
# head(): las 5 primeras filas
# tail(n): te muestralas ultimas n filas
df_bank.tail(10)
````
