````python
# Primera version

def clean_1(dataframe):
  """
  DOCUMENTACIN
  """
  # Generamos una copia de nuestro datadrame
  df_func = dataframe.copy()
  # Rellenamos faltantes
  df_func["Fecha_Fin"] = df_func["Fecha_Fin"].fillna("1900-01-01")
  df_func["Plazo"] = df_bank["df_funcazo"].fillna(0)

  # Homologamos columnas de fecha
  df_func["Fecha_Inicio"] = df_func["Fecha_Inicio"].apply(lambda x: parser.parse(x))
  df_func["Fecha_Fin"] = df_func["df_funcecha_Fin"].apply(lambda x: parser.parse(x))

  # Tipos de datos
  df_func["Plazo"] = df_func["Plazo"].astype(int)

  return df_func
````

````python
# Version mejorada
def limpiezaDatos(df):
    """
    Limpia un archivo CSV de cr茅ditos bancarios con datos sucios.

    Par谩metros:
    ruta_csv (str): Ruta al archivo CSV sucio.

    Retorna:
    pd.DataFrame: DataFrame limpio con fechas unificadas, nombres capitalizados y datos num茅ricos.
    """

    # Funci贸n segura para parsear fechas
    def parsear_fecha(x):
        if pd.isna(x) or str(x).strip().upper() == "N/A":
            return pd.NaT
        try:
            return parser.parse(str(x), dayfirst=True)
        except (parser.ParserError, ValueError):
            return pd.NaT

    # Unificar fechas
    for col in ['Fecha_Inicio', 'Fecha_Fin']:
        df[col] = df[col].apply(parsear_fecha)
        df[col] = pd.to_datetime(df[col], errors='coerce').dt.strftime('%Y-%m-%d')

    # Capitalizar nombres de empresas
    # quitar espacios con strip()
    # Primera letra mayuscula con title()
    df["Nombre_Empresa"] = df["Nombre_Empresa"].str.strip().str.title()

    # Convertir columnas num茅ricas
    df["Monto_Credito"] = pd.to_numeric(df["Monto_Credito"], errors='coerce')
    df["Plazo"] = pd.to_numeric(df["Plazo"], errors='coerce')

    return df
````

````python
from datetime import datetime
import random as r

zonas = ["NORTE", "SUR", "CENTRO"]

# rellenamos con la fecha de hoy
df_bank_clean['Fecha'] = datetime.today().strftime('%Y-%m-%d')

# Agregamos de manera aleatoria la zona
df_bank_clean['Zona'] = [r.choice(zonas) for i in range(len(df_bank_clean))]

# funcion para clasificar al cliente
clasificador = lambda x: "PLUS" if x > 100000000 else "NORMAL"
df_bank_clean['Clasificacion'] = df_bank_clean['Monto_Credito'].apply(clasificador)

# Valorizamos la columna de monto
def valorizacion(x):
  if x == "USD":
    return 18
  elif x == "EUR":
    return 20
  else:
    return 1

df_bank_clean['Valorizacion'] = df_bank_clean['Divisa'].apply(valorizacion)
df_bank_clean['Monto_Valorizado'] = df_bank_clean['Monto_Credito'] * df_bank_clean['Valorizacion']
df_bank_clean.head()
````
