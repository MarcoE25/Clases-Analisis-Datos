### Funci贸n apply de pandas 

* Aplica una funci贸n, idealmente an贸nima, a cada elemento/fila de una columna

````python
# 1. Acceder a la columna de interes
# Fecha_Inicio
df_bank["Fecha_Inicio"] = df_bank["Fecha_Inicio"].apply(lambda x: parser.parse(x))

# fillna(X): te rellena los valores faltantes de la columna de interes
# con el valor X
df_bank["Fecha_Fin"] = df_bank["Fecha_Fin"].fillna("1900-01-01")
df_bank["Fecha_Fin"] = df_bank["Fecha_Fin"].apply(lambda x: parser.parse(x))
````

````python
df_bank["Plazo"] = df_bank["Plazo"].fillna(0)
````

````python
# te da el tipo de dato de tus columnas
# cuantos nulos por columnas tienes
df_bank.info()
````

````python
# conversion de tipos de datos astype(nuevoTipoDeDatos)
df_bank["Plazo"] = df_bank["Plazo"].astype(int)
````
